````python
def lamp_times(df, user_id):
    """
    Use the dataframe obtained from the previous question.
    Return how long a user's lamp has value ON, OFF or DIM.
    The return value should be a tuple (ON, OFF, DIM), where ON, OFF and DIM are of type timedelta.
    """
    #sub df for user_id
    user_info = df[df['user_id'] == user_id].copy()
    #differece between datetimes for every rows
    user_info['diff'] = user_info['datetime'].diff(periods = 1).shift(periods = -1)
    #group by values and sum time
    time = user_info.groupby('value')['diff'].sum().sort_index(ascending = False)

    return tuple(list(time))
````

## `highest_rated_movie_min` (4 points)

````python
def highest_rated_movie_min(ratings, movies, min_n):
    """
    Return the name of the movie with the highest average rating and 
    having at least `min_n` ratings, where `min_n` is an input parameter.
    
    The return value should be the full movie title name including release year,
    e.g., "Toy Story (1995)". This value can be inferred from the 'movie title' 
    column in the movies table.
    """
    #merge
    merged = pd.merge(ratings, movies, left_on = 'item_id', right_on = 'movie id')
    #filter
    min_n_ratings = merged.groupby('movie title').filter(lambda x: len(x) >= min_n)
    
    calculated_rating = min_n_ratings.groupby('movie title')['rating'].agg(['mean','min'])
    return calculated_rating[calculated_rating['mean'] == calculated_rating['mean'].max()].index[0]
````

````python
def mismatched_release_date(movies):
    """
    All movie titles contain the year of release, e.g., Toy Story (1995).
    However, if you pay close attention to some of them, you will notice that
    these years do not always match the values in the "release date" column.
    For example, the last movie is "Scream of Stone (Schrei aus Stein) (1991)"
    whereas the value in the "release date" column is 08-Mar-1996.
    
    Return the number of movies whose release date in the movie title name 
    does not match the date provided in the "release date" column. The returned
    value should be an integer.
    
    Notes
    -----
    - If a movie title name does not contain a release date, or if the release
      date column is empty, then you should ignore the corresponding row entry.
    - Hint: You may assume that every movie title ends with "(YEAR)".
    """
    movies_copy = movies.copy()
    # convert to datetime
    movies_copy["release date"] = pd.to_datetime(movies_copy["release date"])
    title = movies_copy['movie title'].tolist()
    # extra year from title
    movies_copy['title year'] = [int(x[-5:-1]) for x in title]    
    return len(movies_copy[movies_copy["release date"].dt.year != movies_copy['title year']].index)
    
````

## `deadliest_month` (6 points)

````python
def deadliest_month(dfs, state):
    """    
    Return the month during which the most deaths have occured for the passed-in state.
    The month should be an integer value, where 1 denotes January, 2 denotes February, etc.
    
    Notes
    -----
    - The data starts on 1 January 2022 and ends on 9 June 2022. 
    - To get the death count for a certain day, you need to subtract the "Deaths"
      value for two consecutive days. For example, the number of deaths on January 2nd
      can be counted by subtracting the "Deaths" count of January 2nd with January 1st.
    - We do not consider data for 31 December 2021, hence you may assume that there 
      are 0 deaths on January 1st.            
    """
    
    state_data = concat_data(dfs)[concat_data(dfs)['Province_State'] == state].copy()
    state_data['No_death'] = state_data['Deaths'].diff(periods = 1).fillna(0)
    state_data['Date'] = pd.to_datetime(state_data['Date'])
    state_data['Month'] = state_data['Date'].dt.month
    
    return state_data.groupby('Month')['No_death'].sum().sort_values(ascending = False).index[0]
    
````

````python
def daily_number_confirmed(dfs, states):
    """
    The daily reports keep track of the aggregate values for the confirmed
    test cases and deaths. Using this data, it is also possible to infer the 
    newly confirmed cases.
    
    Return the day with the highest number of newly confirmed cases 
    for a list of states. The day should be formatted as, e.g., '01-01-2021', 
    with leading zeros for the day and month. This can be inferred from the
    'Date' column.
    
    Notes (similar to previous question)
    ------------------------------------
    - The data starts on 1 January 2022 and ends on 9 June 2022. 
    - To get the newly confirmed cases count for a given day, you need to subtract the 
      "Confirmed" value for two consecutive days. For example, the number of newly confirmed
      cases on January 2nd can be counted by subtracting the "Confirmed" count of January 2nd
      with January 1st.
    - We do not consider data for 31 December 2021, hence you may assume that there 
      are 0 newly confirmed cases on January 1st.            
    """
    #sum of cf for all states on 1 day
    states_data = concat_data(dfs)[concat_data(dfs)['Province_State'].isin(states)].copy()
    for i in states:
        i_cf = states_data[states_data['Province_State'] == i]['Confirmed'].diff(periods = 1).fillna(0)
        states_data.loc[states_data['Province_State'] == i, 'New_confirmed'] = i_cf.values
    
    return states_data.groupby('Date')['New_confirmed'].sum().sort_values(ascending = False).index[0]
````

## `births_year` (2 points)

````python
def births_year(names, year):
    """
    Return the number of births for a given year.
    """
    birth_per_year = names.groupby('year')['births'].sum()
    return birth_per_year[birth_per_year.index == year].iloc[0]
````

## `difference_female_male_births` (4 points)

````python
def difference_female_male_births(names, year):
    """
    Return the difference in the number of female and male births for a given year.
    
    The difference should be given as:
    diff = (the number of female births) - (the number of male births)
    """
    birth_per_year_male = names[names['sex'] == "M"].groupby('year')['births'].sum()
    birth_per_year_female = names[names['sex'] == "F"].groupby('year')['births'].sum()
    diff = birth_per_year_female - birth_per_year_male
    return diff[diff.index == year].iloc[0]
````

````python
assert_equal(difference_female_male_births(names, 2000), -147960)
````

## `last_letter_popularity` (6 points)

````python
def last_letter_popularity(names, letter):
    """
    According to baby name researcher Laura Wattenberg, the distribution of male names
    has changed significantly over the last years. In particular, male names ending with
    the letter 'n' have become extremely popular in recent decades [1], whereas names
    ending with 'd' and 'y' have become less popular.
    
    Return the fraction of births of males (relative to the total number of male births over all years) 
    that have a name that ends with the passed-in letter.
    
    References
    ----------
    [1] https://qz.com/1278574/a-large-share-of-us-baby-names-end-with-n-for-some-reason/
    """
    total_male_birth = names[names['sex'] == "M"]['births'].sum()
    end_letter = names[(names['sex'] == "M") & (names['name'].str[-1] == letter)]
    return end_letter['births'].sum()/total_male_birth
````

````python
assert_equal(round(last_letter_popularity(names, 'b'), 3), 0.019)
````
